```python
bike_index = bike.set_index(['workingday','mnth']).sort_index()
```

```python
bike_index
```

```python
bike_index1 = bike_index.loc[1] # working days
bike_index2 = bike_index.loc[0] # non working days
```

```python
def plot(bike_index1, xlabel):
    X = bike_index1[['hr', 'weekday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']]
    Y = bike_index1['cnt']
    from sklearn.model_selection import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.33, random_state=42)
    lm = LinearRegression()
    lm.fit(X_train, Y_train)
    predictions = lm.predict(X_test)
    print('predictions: ' + str(predictions[0]))
    print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
    print('MSE:', metrics.mean_squared_error(Y_test, predictions))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
    print(lm.score(X_test, Y_test))
    plt.figure(figsize=(12,2.5))
    plt.scatter(lm.predict(X_train), lm.predict(X_train) - Y_train, c='b', s=40, alpha=0.5)
    plt.scatter(lm.predict(X_test), lm.predict(X_test) - Y_test, c='g', s=40)
    plt.title('Training in blue, test in green')
    plt.ylabel('Residuals')
    plt.xlabel(xlabel)
```

```python
for i in range(1, 13):
    xlabel = 'month ' + str(i)
    bindex = bike_index1
    plot(bindex, xlabel)
```

```python
for i in range(1, 13):
    xlabel = 'month ' + str(i)
    bindex = bike_index2
    plot(bindex, xlabel)
```
